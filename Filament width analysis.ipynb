```python
import os 
import numpy as np
from collections import Counter
from matplotlib import *


#Lists that will contain data for analysis

polygonal_green_pointcount=[]   #note: point count data is enumerated in a dictionary
polygonal_pink_pointcount=[]
flat_green_pointcount=[]
flat_pink_pointcount=[]
polygonal_green_widths=[]
polygonal_pink_widths=[]
flat_green_widths=[]
flat_pink_widths=[]
old_polygonal_pointcount=[]
old_polygonal_widths=[]
old_flat_pointcount=[]
old_flat_widths=[]

#Obtain polygonal mat point count data#

filenames_pol_green_pc=[]
filenames_pol_pink_pc=[]

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if (file.startswith('pol') or file.startswith('Pol') or file.startswith('top')) and file.endswith('PC.txt'):
        if "green" in file or "top" in file:
            filenames_pol_green_pc.append(file)

for i in filenames_pol_green_pc:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        polygonal_green_pointcount.append(x.split('\t')[0])
        filename.close()
        
for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if (file.startswith('pol') or file.startswith('Pol')) and file.endswith('PC.txt'):
        if "pink" in file:
            filenames_pol_pink_pc.append(file)

for i in filenames_pol_pink_pc:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        polygonal_pink_pointcount.append(x.split('\t')[0])
        filename.close()

#Enumerate and sort polygonal mat point count data#

polygonal_green_PC_dict=dict((Counter(polygonal_green_pointcount)))

polygonal_pink_PC_dict=dict((Counter(polygonal_pink_pointcount)))

#Obtain flat mat point count data#

filenames_flat_green_pc=[]
filenames_flat_pink_pc=[]

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if file.startswith('flat') and file.endswith('PC.txt'):
        if "top" in file:
            filenames_flat_green_pc.append(file)

for i in filenames_flat_green_pc:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        flat_green_pointcount.append(x.split('\t')[0])
        filename.close()

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if file.startswith('flat') and file.endswith('PC.txt'):
        if "bottom" in file:
            filenames_flat_pink_pc.append(file)

for i in filenames_flat_pink_pc:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        flat_pink_pointcount.append(x.split('\t')[0])
        filename.close()

#Enumerate and sort flat mat point count data#

flat_green_PC_dict=dict((Counter(flat_green_pointcount)))

flat_pink_PC_dict=dict((Counter(flat_pink_pointcount)))

#Obtain old polygonal mat point count data#

filenames_oldpol_pc=[]

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if file.startswith('old_pol') and file.endswith('PC.txt'):
        filenames_oldpol_pc.append(file)

for i in filenames_oldpol_pc:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        old_polygonal_pointcount.append(x.split('\t')[0])
        filename.close()

#Enumerate and sort old polygonal mat point count data#

old_polygonal_PC_dict=dict((Counter(old_polygonal_pointcount)))

#Obtain old flat mat point count data#

filenames_oldflat_pc=[]

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if file.startswith('old_flat') and file.endswith('PC.txt'):
        filenames_oldflat_pc.append(file)

for i in filenames_oldflat_pc:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        old_flat_pointcount.append(x.split('\t')[0])
        filename.close()

#Enumerate and sort old flat mat point count data#

old_flat_PC_dict=dict((Counter(old_flat_pointcount)))        

#Obtain polygonal mat filament width data#

filenames_pol_green_widths=[]
filenames_pol_pink_widths=[]

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if (file.startswith('pol') or file.startswith('Pol') or file.startswith('top')) and file.endswith('widths.txt'):
        if "green" in file or "top" in file:
            filenames_pol_green_widths.append(file)

for i in filenames_pol_green_widths:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        polygonal_green_widths.append(x.split('\t')[4])
        filename.close()

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if (file.startswith('pol') or file.startswith('Pol')) and file.endswith('widths.txt'):
        if "pink" in file:
            filenames_pol_pink_widths.append(file)

for i in filenames_pol_pink_widths:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        polygonal_pink_widths.append(x.split('\t')[4])
        filename.close()

#Obtain flat mat filament width data#

filenames_flat_green_widths=[]
filenames_flat_pink_widths=[]

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if file.startswith('flat') and file.endswith('widths.txt'):
        if "top" in file:
            filenames_flat_green_widths.append(file)

for i in filenames_flat_green_widths:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        flat_green_widths.append(x.split('\t')[4])
        filename.close()

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if file.startswith('flat') and file.endswith('widths.txt'):
        if "bottom" in file:
            filenames_flat_pink_widths.append(file)

for i in filenames_flat_pink_widths:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        flat_pink_widths.append(x.split('\t')[4])
        filename.close()

#Obtain old polygonal mat filament width data#

filenames_oldpol_widths=[]

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if file.startswith('old_pol') and file.endswith('widths.txt'):
        filenames_oldpol_widths.append(file)
        
for i in filenames_oldpol_widths:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        old_polygonal_widths.append(x.split('\t')[4])
        filename.close()
        
#Obtain old flat mat filament width data#

filenames_oldflat_widths=[]

for file in os.listdir(os.chdir('/Users/sydneyriemer/Desktop/Jmicrovision data')):
    if file.startswith('old_flat') and file.endswith('widths.txt'):
        filenames_oldflat_widths.append(file)
        
for i in filenames_oldflat_widths:
    filename=open(i,'r') 
    lines=filename.readlines()[1:]
    for x in lines:
        old_flat_widths.append(x.split('\t')[4])
        filename.close()
```

```python
polygonal_widths=[]
for i in polygonal_green_widths:
    polygonal_widths.append(i)
for i in polygonal_pink_widths:
    polygonal_widths.append(i)
polygonal_widths=[float(i) for i in polygonal_widths]

flat_widths=[]
for i in flat_green_widths:
    flat_widths.append(i)
for i in flat_pink_widths:
    flat_widths.append(i)
flat_widths=[float(i) for i in flat_widths]

polygonal_green_widths=[float(i) for i in polygonal_green_widths]
polygonal_pink_widths=[float(i) for i in polygonal_pink_widths]
flat_green_widths=[float(i) for i in flat_green_widths]
flat_pink_widths=[float(i) for i in flat_pink_widths]
old_polygonal_widths=[float(i) for i in old_polygonal_widths]
old_flat_widths=[float(i) for i in old_flat_widths]

#gets n for each list
print('n:')
print(len(polygonal_widths))
print(len(polygonal_green_widths))
print(len(polygonal_pink_widths))
print(len(flat_widths))
print(len(flat_green_widths))
print(len(flat_pink_widths))
print(len(old_polygonal_widths))
print(len(old_flat_widths))

#calculates average
polygonal_widths_mean=np.mean(polygonal_widths)
polygonal_green_widths_mean=np.mean(polygonal_green_widths)
polygonal_pink_widths_mean=np.mean(polygonal_pink_widths)
flat_widths_mean=np.mean(flat_widths)
flat_green_widths_mean=np.mean(flat_green_widths)
flat_pink_widths_mean=np.mean(flat_pink_widths)
old_polygonal_widths_mean=np.mean(old_polygonal_widths)
old_flat_widths_mean=np.mean(old_flat_widths)

print('mean:')
print(polygonal_widths_mean)
print(polygonal_green_widths_mean)
print(polygonal_pink_widths_mean)
print(flat_widths_mean)
print(flat_green_widths_mean)
print(flat_pink_widths_mean)
print(old_polygonal_widths_mean)
print(old_flat_widths_mean)

#calculates standard deviation
polygonal_widths_std=np.std(polygonal_widths)
polygonal_green_widths_std=np.std(polygonal_green_widths)
polygonal_pink_widths_std=np.std(polygonal_pink_widths)
flat_widths_std=np.std(flat_widths)
flat_green_widths_std=np.std(flat_green_widths)
flat_pink_widths_std=np.std(flat_pink_widths)
old_polygonal_widths_std=np.std(old_polygonal_widths)
old_flat_widths_std=np.std(old_flat_widths)

print('standard deviation:')
print(polygonal_widths_std)
print(polygonal_green_widths_std)
print(polygonal_pink_widths_std)
print(flat_widths_std)
print(flat_green_widths_std)
print(flat_pink_widths_std)
print(old_polygonal_widths_std)
print(old_flat_widths_std)

#calculates variance
polygonal_widths_var=np.var(polygonal_widths)
polygonal_green_widths_var=np.var(polygonal_green_widths)
polygonal_pink_widths_var=np.var(polygonal_pink_widths)
flat_widths_var=np.var(flat_widths)
flat_green_widths_var=np.var(flat_green_widths)
flat_pink_widths_var=np.var(flat_pink_widths)
old_polygonal_widths_var=np.var(old_polygonal_widths)
old_flat_widths_var=np.var(old_flat_widths)

print('variance:')
print(polygonal_widths_var)
print(polygonal_green_widths_var)
print(polygonal_pink_widths_var)
print(flat_widths_var)
print(flat_green_widths_var)
print(flat_pink_widths_var)
print(old_polygonal_widths_var)
print(old_flat_widths_var)
```

```python
#Whole range t-tests
import scipy
from scipy import stats


print('1', stats.ttest_ind(polygonal_green_widths, polygonal_pink_widths,equal_var=False))
print('2',stats.ttest_ind(polygonal_green_widths, flat_green_widths,equal_var=False))
print('3',stats.ttest_ind(polygonal_green_widths, flat_pink_widths,equal_var=False))
print('4',stats.ttest_ind(polygonal_pink_widths, flat_pink_widths,equal_var=False))
print('5',stats.ttest_ind(polygonal_pink_widths, flat_green_widths,equal_var=False))
print('6',stats.ttest_ind(flat_green_widths, flat_pink_widths,equal_var=False))
print('7',stats.ttest_ind(polygonal_widths,flat_widths,equal_var=False)) #hypothesis about cyanobacteria diversity btwn flat and pol
print('8',stats.ttest_ind(old_flat_widths, old_polygonal_widths,equal_var=False))
print('9',stats.ttest_ind(old_flat_widths, polygonal_widths,equal_var=False))
print('10',stats.ttest_ind(old_flat_widths, flat_widths,equal_var=False))
print('11',stats.ttest_ind(old_flat_widths, flat_green_widths,equal_var=False))
print('12',stats.ttest_ind(old_flat_widths, flat_pink_widths,equal_var=False))
print('13',stats.ttest_ind(old_flat_widths, polygonal_green_widths,equal_var=False))
print('14',stats.ttest_ind(old_flat_widths, polygonal_pink_widths,equal_var=False))
print('15',stats.ttest_ind(old_polygonal_widths, polygonal_widths,equal_var=False))
print('16',stats.ttest_ind(old_polygonal_widths, flat_widths,equal_var=False))
print('17',stats.ttest_ind(old_polygonal_widths, flat_green_widths,equal_var=False))
print('18',stats.ttest_ind(old_polygonal_widths, flat_pink_widths,equal_var=False))
print('19',stats.ttest_ind(old_polygonal_widths, polygonal_green_widths,equal_var=False))
print('20',stats.ttest_ind(old_polygonal_widths, polygonal_pink_widths,equal_var=False))
```
